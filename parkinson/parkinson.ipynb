```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer

def engineer_features(df):
    # Age-related features
    df['age_squared'] = df['age'] ** 2
    df['age_risk'] = pd.qcut(df['age'], q=5, labels=[1,2,3,4,5])
    
    # Blood pressure features
    df['bp_category'] = pd.cut(df['trestbps'], 
                              bins=[0, 120, 140, 160, 999], 
                              labels=['normal', 'prehypertension', 'stage1', 'stage2'])
    df['pulse_pressure'] = df['trestbps'] - df['thalach']
    
    # Heart rate reserve
    df['max_hr'] = 220 - df['age']
    df['hr_reserve'] = df['max_hr'] - df['thalach']
    df['hr_reserve_ratio'] = df['thalach'] / df['max_hr']
    
    # Cholesterol ratios and categories
    df['chol_risk'] = pd.qcut(df['chol'], q=4, labels=['low', 'moderate', 'high', 'very_high'])
    
    # Combined risk factors
    df['cardiovascular_load'] = df['trestbps'] * df['hr_reserve_ratio']
    df['risk_score'] = (df['age_risk'].astype(int) + 
                       df['trestbps']/100 + 
                       df['chol']/200 + 
                       (1 if df['fbs'] > 120 else 0))
    
    # Exercise-related features
    df['exercise_angina'] = df['exang'] * df['oldpeak']
    df['stress_index'] = df['oldpeak'] * df['slope']
    
    # Multiple condition indicators
    df['multiple_conditions'] = ((df['trestbps'] > 140).astype(int) + 
                               (df['chol'] > 200).astype(int) + 
                               (df['fbs'] > 120).astype(int))
    
    # ECG-based features
    df['ecg_stress_indicator'] = df['restecg'] * df['oldpeak']
    
    # Vessel and perfusion features
    df['vessel_score'] = df['ca'] * (4 - df['thal'])
    
    # Normalize new numerical features
    scaler = StandardScaler()
    numerical_features = ['age_squared', 'pulse_pressure', 'hr_reserve', 
                         'cardiovascular_load', 'risk_score', 'exercise_angina', 
                         'stress_index', 'vessel_score']
    df[numerical_features] = scaler.fit_transform(df[numerical_features])
    
    # Encode categorical features
    categorical_features = ['bp_category', 'chol_risk']
    df = pd.get_dummies(df, columns=categorical_features)
    
    return df

def select_features(df, target_col='target'):
    from sklearn.feature_selection import SelectFromModel
    from lightgbm import LGBMClassifier
    
    X = df.drop(target_col, axis=1)
    y = df[target_col]
    
    selector = SelectFromModel(LGBMClassifier(random_state=42, n_jobs=-1), 
                             prefit=False)
    selector.fit(X, y)
    
    selected_features = X.columns[selector.get_support()].tolist()
    return df[selected_features + [target_col]]
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek
import optuna

def create_advanced_knn(x_train, x_test, y_train, y_test):
    # Create feature selector using Random Forest
    selector = SelectFromModel(
        RandomForestClassifier(n_estimators=100, random_state=42),
        max_features=8
    )
    
    # Create pipeline with SMOTE for class imbalance
    pipeline = ImbPipeline([
        ('scaler', RobustScaler()),
        ('feature_selector', selector),
        ('sampler', SMOTETomek(random_state=42)),
        ('classifier', KNeighborsClassifier())
    ])
    
    # Define hyperparameter search space
    param_grid = {
        'classifier__n_neighbors': range(3, 21, 2),
        'classifier__weights': ['uniform', 'distance'],
        'classifier__metric': ['euclidean', 'manhattan', 'minkowski'],
        'classifier__p': [1, 2, 3],
        'classifier__leaf_size': [10, 20, 30, 40]
    }
    
    # Optuna optimization for hyperparameters
    def objective(trial):
        n_neighbors = trial.suggest_int('n_neighbors', 3, 20)
        weights = trial.suggest_categorical('weights', ['uniform', 'distance'])
        metric = trial.suggest_categorical('metric', ['euclidean', 'manhattan', 'minkowski'])
        p = trial.suggest_int('p', 1, 3)
        leaf_size = trial.suggest_int('leaf_size', 10, 40)
        
        knn = KNeighborsClassifier(
            n_neighbors=n_neighbors,
            weights=weights,
            metric=metric,
            p=p,
            leaf_size=leaf_size
        )
        
        pipeline.set_params(classifier=knn)
        scores = cross_val_score(pipeline, x_train, y_train, cv=5)
        return scores.mean()
    
    study = optuna.create_study(direction='maximize')
    study.optimize(objective, n_trials=100)
    
    # Get best parameters
    best_params = study.best_params
    
    # Create final model with best parameters
    final_knn = KNeighborsClassifier(
        n_neighbors=best_params['n_neighbors'],
        weights=best_params['weights'],
        metric=best_params['metric'],
        p=best_params['p'],
        leaf_size=best_params['leaf_size']
    )
    
    final_pipeline = ImbPipeline([
        ('scaler', RobustScaler()),
        ('feature_selector', selector),
        ('sampler', SMOTETomek(random_state=42)),
        ('classifier', final_knn)
    ])
    
    # Fit and evaluate
    final_pipeline.fit(x_train, y_train)
    train_score = final_pipeline.score(x_train, y_train)
    test_score = final_pipeline.score(x_test, y_test)
    
    return final_pipeline, train_score, test_score, best_params

# Usage
model, train_acc, test_acc, best_params = create_advanced_knn(x_train, x_test, y_train, y_test)
print(f"Best Parameters: {best_params}")
print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```
